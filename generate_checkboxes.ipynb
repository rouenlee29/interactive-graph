```python
import json
import numpy as np
```

```python
with open('data-v0.1.json', encoding='utf-8') as data_file:
    data = json.loads(data_file.read())
```

```python
nodes = data["nodes"]
movie_names = []
for d in nodes:
    # collect movie names
    movie_name = d["id"]
    movie_names.append(movie_name)
sorted_idx = np.argsort(np.array(movie_names))
```

```python
for s in sorted_idx:
    idx = nodes[s]["idx"]
    movie_name = nodes[s]["id"]
    html_str = f"""
    <input type="checkbox" name="movie{idx}" value=idx>
    <label for="movied{idx}">{movie_name}</label><br>
    """
    print(html_str)
```
